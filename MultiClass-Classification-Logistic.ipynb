```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading test and train data

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

If we observe the above data column 'id' has no role to play and unnecesarily taking the space so we can drop the column.
Also, the target column is non-categorical so we need to convert it into categorical first.

```python
train_df = train_df.drop('id',axis=1)
test_df = test_df.drop('id',axis=1)
```

Taking target column as y for train data and rest of the feature data as x for the train data

```python
train_y = train_df['target']
train_X = train_df.drop('target',axis=1)
```

```python
train_X.head()
```

Converting train_y into categorical 

```python
for x in range(len(train_y)):
    if train_y[x] == 'Class_1':
        train_y[x] = train_y[x].replace('Class_1','1')
    elif train_y[x] == 'Class_2':
        train_y[x] = train_y[x].replace('Class_2','2')
    elif train_y[x] == 'Class_3':
        train_y[x] = train_y[x].replace('Class_3','3')
    elif train_y[x] == 'Class_4':
        train_y[x] = train_y[x].replace('Class_4','4')
    
```
