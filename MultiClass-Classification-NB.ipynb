Importing required libraries to solve multiclass classification using sklearn

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

Loading train dataset using pandas

```python
train_df = pd.read_csv('train.csv')
```

```python
train_df.head()
```

separating y_train from original train dataset

```python
target = train_df['target']
```

If we can observe now we do not need y labels in train dataframe so dropping that column,and if we can observe column id is not required so dropping it.

```python
train_df = train_df.drop('target',axis=1)
```

```python
train_df = train_df.drop('id',axis=1)
```

```python
train_df.head()
```

Our target colums also y_train labels has alphanumeric values, we can't proceed for any further calculations without changing that.So converting it into numeric

```python
target.head()
# train_df.shape
```

```python
for x in range(100000):
    if target[x] == 'Class_1':
        target[x] = target[x].replace('Class_1','1')
    elif target[x] == 'Class_2':
        target[x] = target[x].replace('Class_2','2')
    elif target[x] == 'Class_3':
        target[x] = target[x].replace('Class_3','3')
    elif target[x] == 'Class_4':
        target[x] = target[x].replace('Class_4','4')
    
```

```python
target
```

Importing test dataset and droping id column

```python
test_df = pd.read_csv('test.csv')
test_df = test_df.drop('id',axis=1)
test_df
```

Using naive bayes algorithm training the classifier and then predicting for test dataset

```python
gnb = GaussianNB().fit(train_df, target)
gnb_predictions = gnb.predict(test_df)
```

```python
print(gnb_predictions)
```

Score using naive bayes

```python
gnb.score(test_df,gnb_predictions)
```

```python
gnb_prob = gnb.predict_proba(test_df)
```

```python
submission = pd.read_csv('sample_submission.csv')
cols = [
    col for col in submission.columns if col != "id"
]
```

```python
submission[cols] = gnb_prob
submission.head()
```
